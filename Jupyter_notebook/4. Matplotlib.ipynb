```python
import numpy as np
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
#
x=np.arange(0.1,4,.01)
plt.figure(figsize=(20,10))                              #
plt.plot(x,np.sin(x*50)/x,linewidth=3)                   #y=sin(50x)/x
plt.plot(x,np.zeros(len(x)),'g--',linewidth=2)           #0'b-'
plt.title('Cureve $y=\sin(50x)/x$',fontsize=40,y=1.04)   #bug};y=1.04
plt.xlabel('$x$',fontsize=30)
plt.ylabel('$y=\sin(50x)/x$',fontsize=30)
plt.grid(True)                                           #yes
plt.savefig('mplsin.png')                                #
```
